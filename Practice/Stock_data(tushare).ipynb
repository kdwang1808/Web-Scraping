```python
import requests
from bs4 import BeautifulSoup
import re
import tushare
```

```python
def getHTMLText(url,header,code="utf-8"):
    try:
        r = requests.get(url,headers=header)
        r.raise_for_status()
        r.encoding = code
        return r.text
    except:
        return ""
```

```python
def getStockList(lst,stockURL):
    kv = {
        "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.97 Safari/537.36"
    }
    html = getHTMLText(stockURL,kv)
    if html != "":
        soup = BeautifulSoup(html,"html.parser")
        td = soup.find_all("td")
        for i in td:
            try:
                #获取股票代码
                code = re.search(r"\d{6}",i.text)     
                if code:
                    code = code.group(0).replace(" ","")
                    lst.append(code)
            except:
                continue
    else:
        print("解析失败")
```

```python
def getStockInfo(lst,fpath):
    count = 0
    infoDict = {}
    for code in lst:
        try:
            dataNow = tushare.get_realtime_quotes(code)
            #股票名字
            name = dataNow.loc[0][0]
            #股票当前价格
            price = float(dataNow.loc[0][3])
 
            with open(fpath,"a",encoding="utf-8") as f:
                f.write(name+"----"+str(price)+"\n")
                count = count + 1
                print("\r当前已经完成:{:.2f}%".format(count*100/len(lst)),end="")
                f.close()
                     
        except:
            count = count + 1
            print("\r当前已经完成:{:.2f}%".format(count*100/len(lst)),end="")
            continue
```

```python
def main():
    stock_list_url = "https://www.txsec.com/inc1/gpdm.asp" 
    #放自己文件地址
    output_file = "C:/Users/Windows10/Desktop/大二.5/Python ML Project/Web Scraping/股票信息.txt" 
    slist = []
    getStockList(slist,stock_list_url)
    getStockInfo(slist,output_file)
```
